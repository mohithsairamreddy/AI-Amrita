Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
```

Basic Graph

```python
x=[1,2,6,8]
y=[3,8,1,10]

## Graph Style
plt.style.use('ggplot')

## Resizing the graph 
plt.figure(figsize=(5,3),dpi=80)

## Label , Line Proporties and marker Properties
plt.plot(x,y,label='line width is 5x',linewidth=5,linestyle="--",color="red",marker='.',markersize=25,markeredgecolor="blue")

## OR plt.plot([1,2,3],[4,5,6])

## Title Parameters and Font parameters
plt.title("Basic Graph",fontdict={"fontname":"sans-serif","fontsize":25},loc="left")

## Adding a label and Font parameters
plt.xlabel("X-Axis",fontdict={"fontname":"monospace'","fontsize":15})
plt.ylabel("Y-Axis",fontdict={"fontname":"monospace'","fontsize":15})

## Adding a legend
plt.legend()

## Adding Grid
plt.grid(color='k')

## To Save graph as png


## To Show Plot(Else will get unnecessary address)
plt.show() 
```

Bar Chat

```python
labels=['C', 'C++', 'Java', 'Python', 'PHP']
values=[23,17,35,29,12]

# Bar Graph
bars= plt.bar(labels,values)


# Title Parameters and Font parameters
plt.title("No of Students vs Programming Language",fontdict={"fontname":"sans-serif","fontsize":15})


# Adding a label and Font parameters
plt.xlabel("Programming Language",fontdict={"fontname":"monospace'","fontsize":15})
plt.ylabel("No of Students",fontdict={"fontname":"monospace'","fontsize":15})

# Adding a legend
plt.legend()


# Patters in Bar Graphs
patters=["/","o","*",".","+"]
for bar in bars:
  bar.set_hatch(patters.pop(0))


# To Show Plot(Else will get unnecessary address)
plt.show()
```

Stack Plot

```python
# List of 7-days
days = [x for x in range(0, 7)]
  
# List of Suspected cases
Suspected = [12, 18, 35, 50, 72, 90, 100]
  
# List of Cured Cases
Cured = [4, 8, 15, 22, 41, 55, 62]
  
# List of Number of deaths
Deaths = [1, 3, 5, 7, 9, 11, 13]

# Plot x-labels, y-label and data
plt.plot([], [], color ='blue', label ='Suspected',linewidth=5)
plt.plot([], [], color ='orange',label ='Cured',linewidth=5)
plt.plot([], [], color ='brown',label ='Deaths',linewidth=5)

 
# There are 4 types of baseline we can use:
# baseline = ["zero", "sym", "wiggle", "weighted_wiggle"]
  
# Implementing stackplot on data
plt.stackplot(days, Suspected, Cured, Deaths, baseline ="sym", colors =['blue', 'orange', 'brown'])

# Adding legend to upper left
plt.legend(loc="upper left")

# Title Parameters and Font parameters
plt.title("No of Cases",fontdict={"fontname":"sans-serif","fontsize":15})

# Adding a label and Font parameters and histogram type
plt.xlabel("Day of the week",fontdict={"fontname":"monospace'","fontsize":15})
plt.ylabel("Overall cases",fontdict={"fontname":"monospace'","fontsize":15})


# To Show Plot(Else will get unnecessary address)
plt.show()
```
